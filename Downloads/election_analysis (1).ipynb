#Nigerian map showing each states won by the four parties 

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Read the dataframe
df = gpd.read_file(r'C:\Users\Mbder\Downloads\nigeriashapefile_NGA_shp\gadm36_NGA_1.shp')

# Define the election results
results = {
    'Abia': 'LP', 'Adamawa': 'PDP', 'Akwa Ibom': 'PDP', 'Anambra': 'LP',
    'Bauchi': 'PDP', 'Bayelsa': 'PDP', 'Benue': 'APC', 'Borno': 'APC',
    'Cross River': 'LP', 'Delta': 'LP', 'Ebonyi': 'LP', 'Edo': 'LP',
    'Ekiti': 'APC', 'Enugu': 'LP', 'Federal Capital Territory': 'LP', 'Gombe': 'PDP', 'Imo': 'LP',
    'Jigawa': 'APC', 'Kaduna': 'PDP', 'Kano': 'NNPP', 'Katsina': 'PDP',
    'Kebbi': 'PDP', 'Kogi': 'APC', 'Kwara': 'APC', 'Lagos': 'LP',
    'Nassarawa': 'LP', 'Niger': 'APC', 'Ogun': 'APC', 'Ondo': 'APC',
    'Osun': 'PDP', 'Oyo': 'APC', 'Plateau': 'LP', 'Rivers': 'APC',
    'Sokoto': 'PDP', 'Taraba': 'PDP', 'Yobe': 'PDP', 'Zamfara': 'APC'
}

# Add a column to the GeoDataFrame indicating the party that won in each state
df['winner'] = df['NAME_1'].apply(lambda state: results[state])

# Define the colors for each party
colors = {'APC': 'red', 'PDP': 'blue', 'LP': 'yellow', 'NNPP': 'green'}

# Create a ListedColormap object from the colors dictionary
cmap = ListedColormap(colors.values())

# Plot the map
fig, ax = plt.subplots(figsize=(10, 10))
df.plot(column='winner', cmap=cmap, legend=True, ax=ax)

# Add state names
for idx, row in df.iterrows():
    state_name = row['NAME_1']
    x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
    ax.annotate(state_name, xy=(x, y), xytext=(3, 3), textcoords="offset points", fontsize=8)

ax.set_title('Nigeria States Election Results')
plt.show()
```
